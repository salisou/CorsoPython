#### Dati di esmpio:
##### *corsi disponibili*

```
corsi = {
    10: "Introduzione a Python",
    11: "Data Analytics con Python",
    12: "Sql Server Avanzato"
}

studenti = {
    100: {
        "nome": "Anna",
        "Cognome": "Barco",
        "Classe": "4A"
    },
    101: {
        "nome": "Marco",
        "Cognome": "Marroni",
        "Classe": "C5"
    },
    103: {
        "nome": "Luca",
        "Cognome": "Russo",
        "Classe": "4B"
    }
}
```

Cicli principale: Chiedi all'utente se vuole aggiungere iscrizioni

```
# ================================
# LISTA DELLE ISCRIZIONI
# Ogni iscrizione √® una tupla: (ID_studente, ID_corso)
# ================================
iscrizioni = []

# Ciclo principale
while True:

    # Mostra i corsi disponibili
    print("\nCorsi disponibili:")
    for id_corso, nome in corsi.items():
        print(id_corso, "_", nome)

    # -------------------------------------
    # INPUT: ID dello studente
    # -------------------------------------
    s = input("Inserisci l'ID dello studente (0 per uscire): ")

    if s == "0":
        break   # esce dal ciclo

    if not s.isdigit():  # verifica se √® un numero
        print("‚ùå L'ID dello studente non √® valido (deve essere un numero).")
        continue

    id_studente = int(s)

    # Controlla se lo studente esiste
    if id_studente not in studenti:
        print("‚ùå Studente non trovato!")
        continue

    # -------------------------------------
    # INPUT: ID del corso
    # -------------------------------------
    c = input("Inserisci l'ID del corso a cui iscrivere lo studente: ")

    if not c.isdigit():  # verifica se √® un numero
        print("‚ùå ID corso non valido (deve essere un numero).")
        continue

    id_corso = int(c)

    # Controlla se il corso esiste
    if id_corso not in corsi:
        print("‚ùå Corso non trovato!")
        continue

    # -------------------------------------
    # CONTROLLO DUPLICATI
    # -------------------------------------
    if (id_studente, id_corso) in iscrizioni:
        print("‚ö†Ô∏è Lo studente √® gi√† iscritto a questo corso.")
    else:
        iscrizioni.append((id_studente, id_corso))
        print(f"üéâ Iscrizione avvenuta: {id_studente} ‚Üí {id_corso}")

    # -------------------------------------
    # MOSTRA TUTTE LE ISCRIZIONI
    # -------------------------------------
    print("\nIscrizioni totali:")
    for s_id, c_id in iscrizioni:
        stud = studenti[s_id]    # prendo il dizionario dello studente
        print(f"{stud['nome']} {stud['Cognome']} ‚Üí {corsi[c_id]}")
```
