#### Dati di esmpio:
##### *corsi disponibili*

```
corsi = {
    10: "Introduzione a Python",
    11: "Data Analytics con Python",
    12: "Sql Server Avanzato"
}

studenti = {
    100: {
        "nome": "Anna",
        "Cognome": "Barco",
        "Classe": "4A"
    },
    101: {
        "nome": "Marco",
        "Cognome": "Marroni",
        "Classe": "C5"
    },
    103: {
        "nome": "Luca",
        "Cognome": "Russo",
        "Classe": "4B"
    }
}
```

Cicli principale: Chiedi all'utente se vuole aggiungere iscrizioni

```
# Lista delle iscrizioni: ogni elemento è tuple (id_Studente, id_corso)
iscriozioni = []

while True:
    print("\nCorsi disponibili:")
    for id_Corso, nome  in corsi.items():
        print(id_Corso, "_", nome)

    # input: id_studente
    s = input("Inserisci l'Id dello studente (0) per uscire")
    if s == "0":
        break


    #esce dal ciclo principale
    if not s.isdigit():
        print("L'Id dello studente non valido (Deve essere un numero )")
        continue

    id_studente = int(s)

    # input: dell'id del corso
    c = input("Inserisci l'id del corso a cui scrivere lo studente:").strip()
    print("Id corso non valido (Deve essere un numero )")
    continue

    id_corso = int(c)

    # controllo: se esiste il corso
    if id_corso not in corsi:
        print("Corso non trovato")
        continue

    # controllo: per evitare iscrizioni duplicate
    if (id_studente, id_corso) in iscriozioni:
        print("Lo studente già scritto a questo corso!")
    else:
        iscriozioni.append((id_studente, id_corso))
        print(f"Iscrizione avvenuta  {id_studente} - {id_corso}")

    # al termine mostra tutte le iscrizioni
    print('\nIscrizioni totali:')
    for s, c in iscriozioni:
        print(f"Studente {[studenti[s]]} -> Corso {corsi[c]}")
```
