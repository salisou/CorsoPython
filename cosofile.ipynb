```python
import matplotlib.pyplot as plt
import pandas as pd

# --- DATA -------------------------------------------------------
studenti = {
    100: {"nome": "Anna",  "classe": "4A", "corsi": [10, 11]},
    101: {"nome": "Marco", "classe": "4A", "corsi": [10]},
    102: {"nome": "Sara",  "classe": "4B", "corsi": [11, 12]},
    103: {"nome": "Luca",  "classe": "4B", "corsi": [10]},
    104: {"nome": "Marta", "classe": "4C", "corsi": [12]},
    105: {"nome": "Paolo", "classe": "5A", "corsi": [10, 11]},
    106: {"nome": "Giada", "classe": "5A", "corsi": [11]},
    107: {"nome": "Elisa", "classe": "5B", "corsi": [12]},
    108: {"nome": "Fabio", "classe": "5C", "corsi": [10]},
    109: {"nome": "Enzo",  "classe": "5C", "corsi": [11, 12]}
}

corsi = {
    10: "Matematica",
    11: "Italiano",
    12: "Informatica"
}

voti = {
    100: {10: 8, 11: 7},
    101: {10: 6},
    102: {11: 9, 12: 8},
    103: {10: 7},
    104: {12: 6},
    105: {10: 9, 11: 8},
    106: {11: 7},
    107: {12: 9},
    108: {10: 5},
    109: {11: 8, 12: 7}
}

```

```python
# --- BUILD DATAFRAME -------------------------------------------
records = []

for id_stud, info in studenti.items():
    for corso in info["corsi"]:
        records.append({
            "id": id_stud,
            "nome": info["nome"],
            "classe": info["classe"],
            "corso": corsi[corso],
            "voto": voti[id_stud].get(corso, None)
        })

df = pd.DataFrame(records)
```

```python

# --- 1. BAR CHART voti medi per studente ------------------------
media_voti = df.groupby("nome")["voto"].mean()

plt.figure(figsize=(10,5))
plt.bar(media_voti.index, media_voti.values)
plt.xticks(rotation=45)
plt.title("Voto medio per studente")
plt.ylabel("Voto medio")
plt.tight_layout()
plt.show()
```

```python
# --- 2. BAR CHART voti per corso -------------------------------
media_corsi = df.groupby("corso")["voto"].mean()

plt.figure(figsize=(8,5))
plt.bar(media_corsi.index, media_corsi.values)
plt.xticks(rotation=45)
plt.title("Voti medi per corso")
plt.ylabel("Voto medio")
plt.tight_layout()
plt.show()
```

```python
# --- 3. TOP 5 STUDENTI -----------------------------------------
top5 = media_voti.sort_values(ascending=False).head(5)

plt.figure(figsize=(10,5))
plt.bar(top5.index, top5.values)
plt.xticks(rotation=45)
plt.title("Top 5 studenti per voto medio")
plt.ylabel("Voto medio")
plt.tight_layout()
plt.show()
```

```python

# --- 4. DISTRIBUZIONE VOTI --------------------------------------
plt.figure(figsize=(8,5))
plt.hist(df["voto"])
plt.title("Distribuzione dei voti")
plt.xlabel("Voto")
plt.ylabel("Frequenza")
plt.tight_layout()
plt.show()
```

```python
# --- 5. STUDENTI CON ALMENO UN CORSO E VOTO > X -----------------
X = 7
df_filtrato = df[df["voto"] > X]

conteggio = df_filtrato.groupby("nome")["voto"].count()

plt.figure(figsize=(10,5))
plt.bar(conteggio.index, conteggio.values)
plt.xticks(rotation=45)
plt.title(f"Studenti con almeno un voto > {X}")
plt.ylabel("Numero voti sopra soglia")
plt.tight_layout()
plt.show()
```
