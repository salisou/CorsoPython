### File Json
### Legge un file JSON

```python
import json
```

```python
studente = {
    "id": 100,
    "nome": "Ilaria",
    "cognome": "Colombraro",
    "classe": "4A",
    "corsi": [10, 11]
}
```

```python
with open("studente.json", "w", encoding="utf-8") as f:
    json.dump(studente, f, indent=5)
```

```python
import json

with open("studente.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(data)
print(data["nome"])
```

### Salvare dati in un file Excel

```python
import pandas as pd
```

```python
dati = {
    "StudenteId": [100, 101, 102],
    "Nome": ["Anna", "Marco", "Luca"],
    "Classe": ["A4", "C5", "4B"]
}
```

```python
df = pd.DataFrame(dati)
print(df)
```

```python
df.to_excel("studenti.xlsx", index=False)
```

### Leggere il contenuto del foglio Excel

```python
import pandas as pd

df = pd.read_excel("studenti.xlsx")
print(df)
```
